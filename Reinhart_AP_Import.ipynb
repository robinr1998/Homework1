```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepaths into variable
reinhart_df = Path("/content/CED_AERO_0202_09022400262362.txt")
```

```python
# Read in files.
reinhart_df = pd.read_csv(reinhart_df)

# Display the DataFrame
reinhart_df.head()
```

```python
reinhart_df.columns
```

```python
# Create ap_import_df
data = {
    "DONOTIMPORT": [],
    "LINE_NO": [],
    "BATCH_TITLE": [],
    "ACCT": [],
    "LOCATION_ID": [],
    "RESTAURANT_NAME_BY_AIRPORT": [],
    "VENDOR_ID": [],
    "BILL_NO": [],
    "CREATED_DATE": [], ## Invoice Date
    "AMOUNT": [],
    "MEMO": [],
    "POSTING_DATE": [],
    "ACCT_NO": []
}
ap_import_df = pd.DataFrame(data)

ap_import_df['LINE_NO'] = ap_import_df['LINE_NO'].astype(int)
#ap_import_df['ACCT_NO'] = ap_import_df['ACCT_NO'].astype(str)
ap_import_df['ACCT'] = ap_import_df['ACCT'].astype(str)
```

```python
# Update ap_import_df DataFrame with values
ap_import_df["BATCH_TITLE"] = "Reinhart Import"
ap_import_df["ACCT"] = reinhart_df['ACCT']
ap_import_df.loc[ap_import_df["ACCT"] == 2592, "LOCATION_ID"] = "DSM_04_PORTERMILL"
#ap_import_df.loc[ap_import_df["ACCT"] == 2592, "LOCATION_ID"] = "DSM_04_PORTERMILL"
#ap_import_df.loc[ap_import_df["ACCT"] == 2592, "LOCATION_ID"] = "DSM_04_PORTERMILL"
ap_import_df["RESTAURANT_NAME_BY_AIRPORT"] = "DSM - Big 5"
#ap_import_df["RESTAURANT_NAME_BY_AIRPORT"] = "DSM - Freddy's"
#ap_import_df["RESTAURANT_NAME_BY_AIRPORT"] = "DSM - Berk"
ap_import_df["VENDOR_ID"] = "V00197" # Reinhart DSM
ap_import_df["BILL_NO"] = reinhart_df["INV #"]
ap_import_df["CREATED_DATE"] = reinhart_df["INV DATE"]
ap_import_df["AMOUNT"] = reinhart_df["EXT PRC"] + reinhart_df["TAX"]
ap_import_df["MEMO"] = reinhart_df["PROD DESCRIPTION"]
ap_import_df["POSTING_DATE"] = reinhart_df["INV DATE"]
ap_import_df["ACCT_NO"] = reinhart_df["GL CODE"].str[:4]

# Add auto numbering
ap_import_df['LINE_NO'] = ap_import_df.groupby('BILL_NO').cumcount() + 1
```

```python
# Drop columns
ap_import_df.drop(columns=['ACCT'], inplace=True)
```
